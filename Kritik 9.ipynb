```python
#question 4b
import sympy as sp
x, y= sp.symbols ('x y')
g=(x**2)*y +x*y**2 # function g 

gradient_vector= [sp.diff(g, variable) for variable in (x,y)]
gradient_point=[derivative.subs({x:1, y:-1}) for derivative in gradient_vector] # for the point (1,-1)
magnitude= sp.sqrt(sum(comp**2 for comp in gradient_point))

print("the gradient vector is", gradient_vector)
print("the gradient at the given point (1,-1) is", gradient_point)
print("the magnetude of the gradient at this point is", magnitude)
```

```python
#question 4c
import sympy as sp
# define some symbols
x, y= sp.symbols ('x y')

#function
h= sp.ln(x**2 + y**2) # i dont know why this print log :( but i gues it doesnt matter cause ln is a specif log

# first partials
dh_dx= sp.diff(h,x) 
dh_dy=sp.diff(h,y)

#second partials
d2h_dx2 = sp.diff(dh_dx,x)
d2h_dy2= sp.diff(dh_dy, y)

#mixed partials
d2h_dyx= sp.diff(dh_dx,y)
d2h_dxy= sp.diff(dh_dy,x)

#just printing it out
print( "The second partial derivative of dh/dx with respect to x is", d2h_dx2)
print()
print( "The second partial derivative of dh/dy with respect to y is", d2h_dy2)
print()
print ("The mixed derivate d2h/dyx is", d2h_dyx)
print()
print( "the mixed derivative d2h_dxy is", d2h_dxy)
```

```python
#Question 4d
#whyyyyyy is there so many functions blah
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define some symbols
x, y= sp.symbols ('x y')
j= x**3 - 3*x*y + y**3

#converting to a numerical function
j_func= sp.lambdify((x,y), j, 'numpy')

#making grid
x_vals =np.linspace(-3, 3, 400)
y_vals= np.linspace(-3, 3, 400)
X, Y =np.meshgrid (x_vals, y_vals)
Z= j_func(X,Y)

# contour plotting stuff
plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x,y)= x^3- 3xy +y$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

```

```python
#question 4d 
#i liked this question
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define some symbols
x, y= sp.symbols ('x y')
e= sp.cos(x)**2 + sp.sin(y)**2

#converting to a numerical function
e_func= sp.lambdify((x,y), e, 'numpy')

#making grid
x_vals =np.linspace(-3, 3, 400)
y_vals= np.linspace(-3, 3, 400)
X, Y =np.meshgrid (x_vals, y_vals)
Z= e_func(X,Y)

# contour plotting stuff
plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $e(x,y)= cos^2(x) + sin^2(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
